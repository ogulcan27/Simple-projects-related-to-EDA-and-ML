MDVP
(Hz):

Mean Fundamental Frequency (Fo): Ortalama temel frekansta ses titreşimlerinin sayısı. Sesin temel frekansını ölçer ve sesin ne kadar hızlı titreştiğini gösterir.

MDVP
(Hz):

Maximum Fundamental Frequency (Fhi): Sesin elde edilebilecek maksimum temel frekansta titreşimlerin sayısı. Sesin en yüksek frekansta ne kadar hızlı titreştiğini gösterir.

MDVP
(Hz):

Minimum Fundamental Frequency (Flo): Sesin elde edilebilecek minimum temel frekansta titreşimlerin sayısı. Sesin en düşük frekansta ne kadar hızlı titreştiğini gösterir.

MDVP
(%):

Jitter (%): Temel frekansta meydana gelen periyodik değişikliklerin yüzdesel olarak hesaplanması. Sesin ne kadar tutarsız olduğunu ölçer.

MDVP
(Abs):

Jitter (Absolute): Temel frekansta meydana gelen periyodik değişikliklerin mutlak değeri. Sesin periyodik değişikliklerini ölçer.

MDVP
:

Relative Average Perturbation (RAP): Temel frekansta meydana gelen periyodik değişikliklerin relatif ortalaması. Sesin periyodik değişikliklerinin ortalamasını hesaplar.

MDVP
:

Pitch Perturbation Quotient (PPQ): Temel frekansta meydana gelen değişikliklerin ölçülmesi. Sesin frekansta meydana gelen değişiklikleri ölçer.

Jitter
:

Jitter: Difference of Differences of Periods (DDP): Periyodik değişikliklerin farklarının farkı. Sesin periyodik değişikliklerinin daha detaylı bir ölçümüdür.

MDVP
:

Shimmer: Amplitüde olan değişikliklerin ölçülmesi. Sesin amplitüdünde meydana gelen değişiklikleri ölçer.

MDVP
(dB):

Shimmer (dB): Amplitüde olan değişikliklerin desibel cinsinden ölçülmesi. Sesin amplitüdündeki değişikliklerin desibel cinsinden ölçümü.

Shimmer
:

Shimmer: Amplitude Perturbation Quotient (APQ3): Amplitüde olan değişikliklerin ortalaması. Sesin amplitüdündeki değişikliklerin ortalamasını ölçer.

Shimmer
:

Shimmer: Amplitude Perturbation Quotient (APQ5): Amplitüde olan değişikliklerin beşinci kısmının ölçülmesi. Sesin amplitüdündeki değişikliklerin ölçümü.

MDVP
:

Amplitude Perturbation Quotient (APQ): Sesin amplitüdündeki değişikliklerin ölçümü. Amplitüd değişikliklerinin genel ölçümüdür.

Shimmer
:

Shimmer: Difference of Differences of Amplitudes (DDA): Amplitüde olan değişikliklerin farklarının farkı. Sesin amplitüdündeki değişikliklerin detaylı bir ölçümüdür.

NHR:

Noise-to-Harmonics Ratio (NHR): Gürültü ile harmonikler arasındaki oran. Sesin gürültü ve harmonik bileşenleri arasındaki oranı ölçer.

HNR:

Harmonics-to-Noise Ratio (HNR): Harmonik bileşenler ile gürültü arasındaki oran. Sesin harmonik bileşenleri ile gürültü arasındaki oranı ölçer.

status:

Parkinson's Disease Status: Parkinson hastalığı durumu (0 veya 1 olarak kodlanmış). 1 genellikle Parkinson hastalığına sahip olduğunu, 0 ise sağlıklı olduğunu gösterir.

RPDE:

Recurrence Plot Density Entropy (RPDE): Sesin karmaşıklığını ve düzensizliğini ölçen bir özellik. Ses sinyalinin tekrar eden desenlerinin yoğunluğunu ölçer.

DFA:

Detrended Fluctuation Analysis (DFA): Ses sinyalindeki uzun dönemli değişiklikleri analiz eder. Sinyalin uzun dönemli varyansını ölçer.

spread1:

Spread 1: Ses sinyalinin yayılma ölçüsü. Sesin dağılımını ölçer.

spread2:

Spread 2: Ses sinyalinin ikinci yayılma ölçüsü. Sesin daha ayrıntılı bir dağılımını ölçer.

D2:

Correlation Dimension (D2): Ses sinyalinin karmaşıklığını ölçen bir özellik. Sesin karmaşıklık düzeyini belirler.

PPE:

Power Spectrum Entropy (PPE): Ses sinyalinin güç spektrumundaki entropi. Sinyalin enerji dağılımının düzensizliğini ölçer.

## PCA (Principal Component Analysis)

```python
pca_14 = project_14
```

```python
pca_14.head(2)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Veriyi standartlaştırma
scaler = StandardScaler()
scaled_data = scaler.fit_transform(pca_14[pca_14.columns].drop(columns = "status", axis = 1))
scaled_data
```

```python
# PCA uygulama
pca = PCA(n_components=3)  # 3 bileşen seçerek başlıyorum
principal_components = pca.fit_transform(scaled_data)
```

```python
# Yeni özellikleri veri çerçevesine ekleme
pca_14['PCA1'] = principal_components[:, 0]
pca_14['PCA2'] = principal_components[:, 1]
pca_14['PCA3'] = principal_components[:, 2]
```

```python
#PCA.. sütunlarının olduğu veri setini döndürme
pca_14.head(3)
```

```python
# Orijinal değişkenleri çıkarma
new_pca_14 = pca_14[["PCA1", "PCA2", "PCA3",  "status"]]
new_pca_14
```

```python
# Yeni VIF hesaplaması
print("\n")
vif_data = pd.DataFrame()
vif_data["Feature"] = new_pca_14.columns
vif_data["VIF"] = [variance_inflation_factor(new_pca_14.values, i) for i in range(len(new_pca_14.columns))]

print(vif_data)
```

### Model Fitting

```python
new_pca_14.head(3)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.model_selection import cross_val_score
```

```python
classification_models = {"Decision Tree": DecisionTreeClassifier(),
                        "Random Forest": RandomForestClassifier(),
                        "Gradient Boosting": GradientBoostingClassifier(),
                        "Ada Boost": AdaBoostClassifier(),
                        "KNN": KNeighborsClassifier(),
                        "Logistic Regression": LogisticRegression(),
                        "Linear SVM": SVC(kernel = "linear"),
                        "RBF SVM": SVC(kernel = "rbf"),
                        "Polynomial SVM": SVC(kernel = "poly")}
```

```python
model_names = []
train_accuracy = []
test_accuracy = []
precision = []
recall = []
f1 = []
roc_auc = []

for name, clf in classification_models.items():
    clf.fit(X_train, y_train)
    predictions = clf.predict(X_test)
    train_score = clf.score(X_train, y_train)
    test_score = accuracy_score(y_test, predictions)
    score_precision = precision_score(y_test, predictions)
    score_recall = recall_score(y_test, predictions)
    score_f1 = f1_score(y_test, predictions)
    score_roc_auc = roc_auc_score(y_test, predictions)

    model_names.append(name)
    train_accuracy.append(train_score)
    test_accuracy.append(test_score)
    precision.append(score_precision)
    recall.append(score_recall)
    f1.append(score_f1)
    roc_auc.append(score_roc_auc)

    print(f"{name} Train setindeki doğruluk: {train_score}")
    print(f"{name} Test setindeki doğruluk: {test_score}")
    print(f"{name} Precision (Hassasiyet): {score_precision}")
    print(f"{name} Recall (Duyarlılık): {score_recall}")
    print(f"{name} F1 Skoru: {score_f1}")
    print(f"{name} ROC-AUC Skoru: {score_roc_auc}\n")
```
