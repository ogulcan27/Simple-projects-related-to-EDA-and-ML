### Hangi bölgede ölüm oranı en yüksek

```python
#Hangi Bölgede Ölüm Oranı En Yüksek
project_4["Death Rate"] = project_4["Deaths"]/project_4["Confirmed"]
print(project_4.head(2), "\n")

project_4.loc[project_4["Death Rate"].idxmax()]["Region"]
```

+ `idxmax()` işlevi, belirtilen sütunda en yüksek değere sahip olan indexi döndürür. Böyle `.loc[]` işlevi ile bu index seçilmiş olur. `["Reigon"]` ifadesi ile Reigon sütununda bu indexe denk gelen gözlem döndürülür

```python
project_4[project_4["Death Rate"] == project_4["Death Rate"].max()]
```
