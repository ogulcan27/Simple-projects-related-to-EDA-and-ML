# Data Cleaning Exercises 1 - Customer Call List

## Veri setini içe aktaralım ve görüntüleyelim

```python
import pandas as pd 

customer_call_list = pd.read_excel("Customer Call List.xlsx")
customer_call_list
```

## İçe aktardığımız veri seti ile ilgili bilgi edinelim

Pandas'ta veri setiyle ilgili bilgi edinmek, sütunların veri tipini öğrenmek için `pandas.DataFrame.info()` işlevini kullanırız.

```python
customer_call_list.info()
```

Veri setinin kaç satır ve sütundan oluştuğunu görebilmek için `pandas.DataFrame.shape` işlevini kullanırız.

```python
customer_call_list.shape
```

## Tekrar eden değerleri bulalım ve kaldıralım

Pandas'ta tekrar eden değerleri bulabilmek için `pandas.DataFrame.duplicated()` işlevini kullanırız.

```python
customer_call_list.duplicated()
```

Görüleceği üzere 20. satır **True** olarak döndürüldü. Demek ki bu satır gereksiz ve tekrar ediyor. Bu satırı `pandas.DataFrame.drop_duplicates` ile bırakalım.

```python
customer_call_list = customer_call_list.drop_duplicates()
customer_call_list
```

Görüleceği üzere tekrar eden 20. satır kaldırılmış oldu.

## İhtiyacımız olmayan sütunları kaldıralım 

Veri setindeki sütunlara `pandas.DataFrame.columns` ile göz atalım.

```python
customer_call_list.columns
```

Bu sütunlar arasında **Not_Useful_Column** işimize yaramayacak. Bu sütunu `pandas.DataFrame.drop()` işlevi ile bırakalım. 

```python
customer_call_list = customer_call_list.drop(columns = "Not_Useful_Column")
customer_call_list
```

Görüleceği üzere, işimize yaramayacak olan sütun kaldırılmış oldu.

## Last_Name sütunudaki yazım hataları ve noktalama yanlışlıklarını temizleyelim 

```python
customer_call_list["Last_Name"]
```

Bir sütundaki yazım hataları ve noktalama yanlışlıklarını gidermek için `str.strip()` işlevini kullanırız. Eğer metnin sol tarafındaki yanlışlığı gidermek istiyorsak `str.lstrip()`, sağ tarafındaki yanlışlığı gidermek istiyorsak `str.rstrip()` işlevini kullanırız. Her iki taraf için direkt olarak `str.strip()` işlevini kullanırız.  

```python
customer_call_list["Last_Name"] = customer_call_list["Last_Name"].str.lstrip("...")
customer_call_list["Last_Name"] = customer_call_list["Last_Name"].str.lstrip("/")
customer_call_list["Last_Name"] = customer_call_list["Last_Name"].str.rstrip("_")
customer_call_list
```

Görüleceği üzere, Potter'ın sol tarafındaki **' ... '**, White'ın sol tarafındaki **' / '**, Flenderson'un sağ tarafındaki **' _ '** gitti. 

## Phone_Number sütununu temizleyelim ve standartlaştıralım  

```python
customer_call_list["Phone_Number"]
```

Telefon numaralarının **123-545-5421** formatında görünmesini istiyoruz. Buna göre düzenleme yapacağız. Bunu yapmak için `replace()` işlevini: `customer_call_list["Phone_Number"].str.replace('[^a-zA-Z0-9]', '')` şeklinde kullanacağız. Bu ifade, "Phone_Number" sütununda bulunan değerlerdeki tüm olmayan harf ve rakamları ("a-z", "A-Z" ve "0-9" dışındaki karakterler) boşluk karakteriyle değiştirir.

```python
customer_call_list["Phone_Number"] = customer_call_list["Phone_Number"].str.replace('[^a-zA-Z0-9]', '')
customer_call_list["Phone_Number"]
```

```python
customer_call_list
```

Daha sonra **Phone_Number** sütununa `apply` işlevinde kullanacağımız bir lambda fonksiyonu ile sütundaki bütün değerleri string veri tipine çevireceğiz. 

```python
customer_call_list["Phone_Number"] = customer_call_list["Phone_Number"].apply(lambda x: str(x))
customer_call_list["Phone_Number"]
```

Ardından `apply()` işlevi ile lambda fonksiyonu kullanarak sütundaki her satıra uygulanacak bir fonksiyon ile her değere, 3 karakterden sonra **' - '** işaretini yerleştireceğiz 

```python
customer_call_list["Phone_Number"] = customer_call_list["Phone_Number"].apply(lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:10])
customer_call_list["Phone_Number"]
```

```python
customer_call_list
```

Daha sonra, **nan--** ve **Na--** olan değerleri `replace()` işlevini kullanarak boşluk ile  değiştireceğiz.

```python
customer_call_list["Phone_Number"] = customer_call_list["Phone_Number"].str.replace("nan--","")
customer_call_list["Phone_Number"] = customer_call_list["Phone_Number"].str.replace("Na--","")
customer_call_list["Phone_Number"]
```

```python
customer_call_list
```

## Address sütunundaki değerleri bölelim

**Address** sütunundaki değerlerin **' , '**'lerine kadarki kısımlarını `str.split()` işlevi ile böleceğiz. 

```python
customer_call_list[["Street_Address", "State", "Zip_Code"]] = customer_call_list["Address"].str.split("," , 2, expand = True)
customer_call_list
```

```python
#Ayrılan Parçalar
customer_call_list["Address"].str.split("," , 2, expand = True)
```

**customer_call_list[["Street_Adress", "State", "Zip_Code"]] = customer_call_list["Address"].str.split("," , 2, expand = True)** kodundaki **2**, `maxsplit` parametresine girilen bir değerdir. Bu parametre, metnin kaç parçaya bölüneceğini söyler. Bu parametreye 2 girildiyse, birden çok virgül olsa dahi sadece ilk 2 virgülden önceki ve sonraki kısımları ayırır. Yani her satırdaki metni 3 parçaya ayırmış olur. Ayrıca bu kod, her parçaya sırasıyla **Street_Address, State ve Zip_Code** isimlerini verir ve 3 yeni sütun oluşturmuş olur.

## Paying Customer ve Do_Not_Contact sütunlarındaki değerleri düzenleyelim

```python
customer_call_list["Paying Customer"]
```

Bu sütundaki **Yes** ve **No** değerlerini, **Y** ve **N** olacak şekilde değiştirelim

```python
customer_call_list["Paying Customer"] = customer_call_list["Paying Customer"].str.replace("Yes", "Y")
customer_call_list["Paying Customer"] = customer_call_list["Paying Customer"].str.replace("No", "N")
customer_call_list["Paying Customer"]
```

Aynı şeyi **Do_Not_Contact** sütunu için uygulayalım.

```python
customer_call_list["Do_Not_Contact"] = customer_call_list["Do_Not_Contact"].replace({"Yes": "Y", "No": "N"})
customer_call_list["Do_Not_Contact"]
```

```python
customer_call_list
```

Paying Customer ve Do_Not_Contact sütunundaki **Null** değerleri boşluk ile değiştirelim 

```python
customer_call_list["Paying Customer"] = customer_call_list["Paying Customer"].replace({"N/a":""})
customer_call_list["Paying Customer"]
```

```python
customer_call_list["Do_Not_Contact"] = customer_call_list["Do_Not_Contact"].replace({"NaN":""})
customer_call_list["Do_Not_Contact"]
```

```python
customer_call_list
```

## Veri setindeki Null değerleri boşluk ile dolduralım 

```python
customer_call_list = customer_call_list.fillna("")
customer_call_list
```

## Do_Not_Contact sütununa göre düzenleme yapalım 

Do_Not_Contact sütunu iletişim sağlanabilen kişileri gösterir. İletişimin sağlanamadığı kişiler bizim işimize yaramayacak. Bu yüzden bu kişileri veri setinden sileceğiz. Ayrıca iletişim sağlanıp sağlanmadığı bilinmeyen kişiler de işimize yaramaz. Bu yüzden bunların olduğu satırları da bırakacağız.

```python
for x in customer_call_list.index:
    if customer_call_list.loc[x, "Do_Not_Contact"] == "Y":
        customer_call_list.drop(x, inplace = True)
        
    elif customer_call_list.loc[x, "Do_Not_Contact"] == "":
        customer_call_list.drop(x, inplace = True)

customer_call_list
```

## Index'leri düzenleyelim 

```python
customer_call_list = customer_call_list.reset_index(drop = True)
customer_call_list
```

## İletişimin sağlandığı ancak telefon numarası olmayan kişilere ulaşım sağlanamayacağından, bu kişilerin olduğu satırları da bırakalım

```python
for x in customer_call_list.index:        
    if customer_call_list.loc[x, "Phone_Number"] == "":
        customer_call_list.drop(x, inplace = True)
        
customer_call_list
```
