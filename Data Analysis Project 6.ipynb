## Q.7) Calculate the gender ratio for each state (Male to Female ratio)

```python
project_6.head()
```

```python
project_6["Gender_ratio"] = project_6["Male"] / project_6["Female"]
project_6.head(2)
```

```python
project_6.groupby("State_name")["Gender_ratio"].sum()
```

```python
gender_ratio = project_6.groupby('State_name').apply(lambda x: x['Male'].sum() / x['Female'].sum())
gender_ratio
```

## Q.10) Find the state with the highest percentage of literate population

```python
project_6.head(2)
```

```python
project_6[project_6["Literacy_rate"] == project_6["Literacy_rate"].max()]
```

```python
state_literacy_rate = project_6.groupby('State_name').apply(lambda x: x['Literate'].sum() / x['Population'].sum() * 100)
highest_literacy_state = state_literacy_rate.idxmax()
highest_literacy_state
```
