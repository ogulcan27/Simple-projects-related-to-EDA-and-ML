## MSRP Tahmini

```python
#Y Bağımlı Değişkeni = MSRP
#X Bağımsız Değişkenleri = MSRP Harici Numerik Sütunlar
X = project_2.drop(["MSRP", "Make", "Model", "Type", "Origin", "DriveTrain"], axis = 1)
y = project_2["MSRP"]
```

```python
#Train ve Test Setlerini Ayırma
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

### Lineer Regresyon

```python
#Lineer Regresyon modelinin oluşturulması
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()

##Algoritmayı Eğitme
linreg.fit(X_train, y_train)

##Modelin Katsayıları
print("Lineer Regresyon modelinin katsayıları:", linreg.coef_)
print("\nLineer Regresyon modelinin sabiti:", linreg.intercept_)

##Sonuçları Yazdırma
print("\nTrain Setindeki Doğruluk:", linreg.score(X_train, y_train))
print("Test Setindeki Doğruluk:", linreg.score(X_test, y_test))

##Yeni veri için tahmin
msrp_prediction = linreg.predict([[25.000, 3.5, 5.0, 250.0, 23.0, 30.0, 4000.0, 110.0, 185.0]])
print(msrp_prediction)
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(linreg, X, y)

print("Her Modeldeki doğruluk oranı:", cv_scores)
print("\nModellerin Doğruluk oranlarının ortalaması:", cv_scores.mean())
```

### Ridge Regresyon

```python
#Ridge Regresyon modelinin oluşturulması
from sklearn.linear_model import Ridge
ridgereg = Ridge()

##Algoritmayı Eğitme
ridgereg.fit(X_train, y_train)

##Modelin Katsayıları
print("Lineer Regresyon modelinin katsayıları:", ridgereg.coef_)
print("\nLineer Regresyon modelinin sabiti:", ridgereg.intercept_)

##Sonuçları Yazdırma
print("\nTrain Setindeki Doğruluk:", ridgereg.score(X_train, y_train))
print("Test Setindeki Doğruluk:", ridgereg.score(X_test, y_test))

##Yeni veri için tahmin
msrp_prediction = ridgereg.predict([[25.000, 3.5, 5.0, 250.0, 23.0, 30.0, 4000.0, 110.0, 185.0]])
print(msrp_prediction)
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(ridgereg, X, y)

print("Her Modeldeki doğruluk oranı:", cv_scores)
print("\nModellerin Doğruluk oranlarının ortalaması:", cv_scores.mean())
```
