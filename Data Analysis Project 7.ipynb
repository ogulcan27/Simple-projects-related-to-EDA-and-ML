## Q.19) What is the average content duration of courses?

```python
project_7.head(2)
```

```python
project_7["content_duration"].unique()
```

```python
def convert_to_minutes(duration):
    if 'hours' in duration:
        return float(duration.replace('hours', '')) * 60

    elif "hour" in duration:
        return float(duration.replace("hour", "")) * 60
    
    elif 'mins' in duration:
        return float(duration.replace('mins', ''))

project_7['duration_minutes'] = project_7['content_duration'].apply(convert_to_minutes)
project_7["duration_minutes"]
```

```python
project_7["duration_minutes"].mean()
```

```python
def convert_to_hours(duration):
    if 'hours' in duration:
        return float(duration.replace('hours', ''))

    elif "hour" in duration:
        return float(duration.replace("hour", ""))
    
    elif 'mins' in duration:
        return float(duration.replace('mins', '')) / 60

project_7['duration_hours'] = project_7['content_duration'].apply(convert_to_hours)
project_7["duration_hours"]
```

```python
project_7["duration_hours"].mean()
```
